```python
import pandas as pd

# 数据没有头
data = pd.read_csv('loss.csv', header=None)
```

```python
data
```

```python
# 科研绘图
import matplotlib.pyplot as plt
import scienceplots
import matplotlib

matplotlib.rcParams['font.family'] = 'Times New Roman'


# 设置绘图风格
plt.style.use(['science', 'no-latex'])

# 绘制折线图
plt.plot(data[0], data[1], label='line')

# 美化图形
plt.title('Iris Classification', fontsize=11)
plt.xlabel('train epoch', fontsize=11)
plt.ylabel('loss', fontsize=11)
## 添加网格
plt.grid(linestyle='--')

# 以dpi=600分辨率保存图片为pdf格式
plt.savefig('loss.pdf', dpi=600)
```
